```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
import random
np.random.uniform(-1, 1, [10, 10])
```

```python
### 1. 請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap
import random
matrix = np.random.uniform(-1, 1, [10, 10])

plt.figure(figsize=(10,10))

sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin=-1, vmax=1, annot=True) 
plt.title('Heat Map of Random Range(-1~1)')
plt.show()
```

```python
### 2. 請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
nrow = 1000
ncol = 3

matrix = np.random.uniform(-1, 1, [nrow,ncol])

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
